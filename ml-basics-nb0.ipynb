```python
import matplotlib.pyplot as plt
%matplotlib inline

# plotting a line plot after changing it's width and height
plt.figure(figsize=(6,6))

import requests
from PIL import Image

def get_image(url):
    headers = {'User-Agent' : 'Mozilla/5.0 ...' }
    response = requests.get(url, stream = True, headers = headers)
    img = Image.open(response.raw)
    return img
```

```python
# Motivation: https://medium.com/ml-research-lab/machine-learning-algorithm-overview-5816a2e6303

plt.axis('off')
plt.imshow(get_image("https://miro.medium.com/v2/resize:fit:1400/format:webp/0*P7pskqJakAFZhF5o.png"))
```
